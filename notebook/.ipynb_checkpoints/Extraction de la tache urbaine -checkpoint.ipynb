```python
image_file = os.path.join("/home/fouzai/Progysat_tache_urbaine/cayenne/images_sentinel2_maja/ACP_synthese_extract.tif")
sample_image = io.imread(image_file)
```

### Utilisation de l'outil FOTOTEX pour calculer les indices de textures : 
Les indices de textures sont calculés sur le canal 1 d'une ACP de l'image RGB.  

On a tourné FOTOTEX avec las paramètres suivants :  
       - méhtode d'analyse = fenetre glissante  
       - taille de la fenetre = 7  
       - dc_component = True  
  
  
voici le lien vers le tutoriel pour l'installation et l'utilisation de FOTOTEX : https://framagit.org/benjaminpillot/fototex

```python

foto_obj = Foto(image_file, method="moving", in_memory=True)
foto_obj.run(window_size=7, keep_dc_component=True)
foto_obj.save_rgb()
```

```python
rgb_image = io.imread(foto_obj.rgb_file)
#rgb_image[rgb_image==-999] = np.nan

# 2%-98% range for values in each channel 
value_min = [-1.42, -1.04, -1.29]
value_max = [6.28, 1.56, 1.22]

for vmin, vmax, band in zip(value_min, value_max, range(rgb_image.shape[2])):
    rgb_image[:,:,band] = (rgb_image[:,:,band] - vmin)/ (vmax - vmin)

plt.figure(figsize=(12,8))
plt.imshow(rgb_image)
plt.axis('off')
```

### Nous allons calculer la variance sur une fenetre glissante  
La taille de la fenetre = 210 X 210 m

```python
import otbApplication

app = otbApplication.Registry.CreateApplication("BandMathX")

app.SetParameterStringList("il", [foto_obj.rgb_file])
app.SetParameterString("out", "/home/fouzai/Progysat_tache_urbaine/cayenne/images_sentinel2_maja/traitement_tu/variance.tif")
app.SetParameterString("exp", "var(im1b1N21x21)")

app.ExecuteAndWriteOutput()
```

La prochaine étape sera consacré sur le seuillage de la variance pour détécter la tache urbaine :

```python
seuil_min = 0.13
seuil_max = 210


import otbApplication

app = otbApplication.Registry.CreateApplication("BandMath")

app.SetParameterStringList("il", ['/home/fouzai/Progysat_tache_urbaine/cayenne/images_sentinel2_maja/traitement_tu/variance.tif'])
app.SetParameterString("out", "/home/fouzai/Progysat_tache_urbaine/cayenne/images_sentinel2_maja/traitement_tu/seuillage_variance.tif")
app.SetParameterString("exp", "((im1b1 > 0.13 && im1b1 < 210) ? 1 : 0)")

app.ExecuteAndWriteOutput()
```

##### On applique des opérations morphologiques sur la variance après le seuillage  ( dilatation de 50 m + érosion de 30 m ) pour extraire la tache urbaine 

```python
import otbApplication

app = otbApplication.Registry.CreateApplication("BinaryMorphologicalOperation")

app.SetParameterString("in", "/home/fouzai/Progysat_tache_urbaine/cayenne/images_sentinel2_maja/traitement_tu/seuillage_variance.tif")
app.SetParameterString("out", "/home/fouzai/Progysat_tache_urbaine/cayenne/images_sentinel2_maja/traitement_tu/seuillage_variance_dilated.tif")
app.SetParameterInt("channel", 1)
app.SetParameterInt("xradius", 5)
app.SetParameterInt("yradius", 5)
app.SetParameterString("filter","dilate")

app.ExecuteAndWriteOutput()

print(" dilatation effectuée avec succès")

app = otbApplication.Registry.CreateApplication("BinaryMorphologicalOperation")

app.SetParameterString("in", "/home/fouzai/Progysat_tache_urbaine/cayenne/images_sentinel2_maja/traitement_tu/seuillage_variance_dilated.tif")
app.SetParameterString("out", "/home/fouzai/Progysat_tache_urbaine/cayenne/images_sentinel2_maja/traitement_tu/tache_urbaine.tif")
app.SetParameterInt("channel", 1)
app.SetParameterInt("xradius", 3)
app.SetParameterInt("yradius", 3)
app.SetParameterString("filter","erode")

app.ExecuteAndWriteOutput()
```
